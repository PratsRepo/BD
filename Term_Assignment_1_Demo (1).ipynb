```python
# libraries
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
 
# Build a dataframe with 5 connections
df = pd.DataFrame({ 'from':['Jean-Luc Picard', 'Zoe Kirk', 'Sarah Smith','Jean-Luc Picard','Zoe Kirk'], 'to':['Zeo Wellington', 'Jean-Luc Picard', 'Robert Mueller','Sarah Smith','Sarah Smith']})
df
```

```python
# Build your graph
G=nx.from_pandas_edgelist(df, 'from', 'to')
 
# Plot it
nx.draw(G, with_labels=True, node_size=8000)
plt.show()
```
